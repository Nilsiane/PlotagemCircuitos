Plotagem de gr√°ficos - Circuitos de primeira ordem.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a fun√ß√£o v(t)
def v(t):
    return 10 - (5 * np.exp(-2 * t))

# Criando os valores de t
t = np.linspace(0, 5, 100)

# Criando os valores de v(t)
vt = v(t)

# Plotando o gr√°fico
plt.plot(t, vt)
plt.xlabel('Tempo (s)')
plt.ylabel('Voltagem (V)')
plt.title('Resposta completa de um circuito RC especifico a uma entrada degrau')
plt.grid()
plt.show()
```

Plotagem Resposta Natural:


```python
# Definindo a fun√ß√£o vn(t)
def vn(t):
    return 5 * np.exp(-2 * t)

# Criando os valores de t
t = np.linspace(0, 5, 100)

# Criando os valores de vn(t)
vnt = vn(t)

# Plotando o gr√°fico
plt.plot(t, vnt)
plt.xlabel('Tempo (s)')
plt.ylabel('Voltagem (V)')
plt.title('Resposta natural')
plt.grid()
plt.show()
```

```python
Podemos observar que a curva da reposta natural se mostrou muito similar 
ao gr√°fico do comportamento do proprio circuito sem nenhuma fonte externa de excita√ß√£o.
```

Resposta For√ßada:

```python
# Definindo a fun√ß√£o vt(t)
def vf(t):
    return 10 - 10 * np.exp(-2 * t)

# Criando os valores de t
t = np.linspace(0, 5, 100)

# Criando os valores de vf(t)
vft = vf(t)

# Plotando o gr√°fico
plt.plot(t, vt)
plt.xlabel('Tempo (s)')
plt.ylabel('Voltagem (V)')
plt.title('Resposta For√ßada')
plt.grid()
plt.show()
```
