Plotagem de gráficos - Circuitos de primeira ordem.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função v(t)
def v(t):
    return 10 - (5 * np.exp(-2 * t))

# Criando os valores de t
t = np.linspace(0, 5, 100)

# Criando os valores de v(t)
vt = v(t)

# Plotando o gráfico
plt.plot(t, vt)
plt.xlabel('Tempo (s)')
plt.ylabel('Voltagem (V)')
plt.title('Resposta completa de um circuito RC especifico a uma entrada degrau')
plt.grid()
plt.show()
```

Plotagem Resposta Natural:


```python
# Definindo a função vn(t)
def vn(t):
    return 5 * np.exp(-2 * t)

# Criando os valores de t
t = np.linspace(0, 5, 100)

# Criando os valores de vn(t)
vnt = vn(t)

# Plotando o gráfico
plt.plot(t, vnt)
plt.xlabel('Tempo (s)')
plt.ylabel('Voltagem (V)')
plt.title('Resposta natural')
plt.grid()
plt.show()
```

```python
Podemos observar que a curva da reposta natural se mostrou muito similar 
ao gráfico do comportamento do proprio circuito sem nenhuma fonte externa de excitação.
```

Resposta Forçada:

```python
# Definindo a função vt(t)
def vf(t):
    return 10 - 10 * np.exp(-2 * t)

# Criando os valores de t
t = np.linspace(0, 5, 100)

# Criando os valores de vf(t)
vft = vf(t)

# Plotando o gráfico
plt.plot(t, vt)
plt.xlabel('Tempo (s)')
plt.ylabel('Voltagem (V)')
plt.title('Resposta Forçada')
plt.grid()
plt.show()
```
